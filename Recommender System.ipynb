```python
import pandas as pd
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
df = pd.read_csv(r"D:\Assignment\book.csv",encoding='latin-1')
```

```python
df1=df.drop('Unnamed: 0', axis=1)
```

```python
df2 =df1.rename({'User.ID':'ID', 'Book.Title':'Title','Book.Rating':'Rating'}, axis=1)
```

```python
df2.isna().sum()
```

```python
df2.sort_values(['ID'])
```

```python
len(df2['ID'].unique())
```

```python
len(df2['Title'].unique())
```

```python
df_piv = df2.pivot_table(values='Rating', index='ID', columns='Title')
```

```python
df_piv
```

```python
df_piv.fillna(0 , inplace= True)
```

```python
df_piv.index = df2['ID'].unique()
```

```python
df_piv
```

```python
df_sim = 1-pairwise_distances(df_piv.values, metric='cosine')
```

```python
df_sim
```

```python
df_sim_array = pd.DataFrame(df_sim)
```

```python
df_sim_array
```

```python
df_sim_array.index = df2['ID'].unique()
df_sim_array.columns = df2['ID'].unique()
df_sim_array
```

```python
np.fill_diagonal(df_sim,0)
df_sim_array
```

```python
def recommend (custID):
    pickID = df_sim_array[custID].idxmax()
    custID_book = df2[df2['ID'] == custID].Title
    pickID_book = df2[df2['ID']==pickID].Title
    return set(pickID_book) -set(custID_book)
```

```python
recommend(276884)
```
